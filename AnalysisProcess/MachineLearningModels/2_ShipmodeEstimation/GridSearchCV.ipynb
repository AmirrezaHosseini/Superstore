```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.multioutput import MultiOutputClassifier as MOC
from sklearn.model_selection import GridSearchCV as GS
```

```python
df = pd.read_csv('dataFrame.csv')
```

```python
X2 = df.drop(['Unnamed: 0', 'Shipping Cost', 'Ship Mode'], axis=1)
X2
```

```python
sc = StandardScaler()
sc.fit(X2)
X2=sc.transform(X2)
```

```python
OneHot = OneHotEncoder()
y = OneHot.fit_transform(df[['Ship Mode']]).toarray()
y
```

```python
param = {
    'estimator__n_estimators': [20, 60, 100, 140,170,190,200],
    'estimator__criterion': ['gini',  'entropy'],
    'estimator__max_depth': [None, 6,7,8,10,12,14,16,18],
    'estimator__min_samples_leaf': [4,6,8,10,12]
}
rfc = RFC()
gsh = GS(MOC(rfc), param_grid=param, scoring='f1', cv=2, n_jobs=-1, verbose=10) 
```

```python
gsh.fit(X2, y)
```

```python
gsh.best_estimator_
```
